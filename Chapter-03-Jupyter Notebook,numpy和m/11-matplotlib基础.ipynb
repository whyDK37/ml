# matplotlib基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x =np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
cosy = np.cos(x)
```

```python
siny = y.copy()
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red", linestyle="--")
plt.show()
```

```python
# 调节坐标
plt.plot(x,siny)
plt.plot(x,cosy,color="red", linestyle="--")
plt.ylim(0,1.5)
plt.show()
```

```python
# axis 同时调节 x，y
# 和其他信息
plt.plot(x,siny,label="sin")
plt.plot(x,cosy, label="cos",color="red", linestyle="--")
plt.axis([-1,11,-2,2])
plt.xlabel("x")
plt.ylabel("value")
plt.legend()
plt.title("welcome")
plt.show()
```

## Scatter Plot 散点图
x,y 一般表示两个特征

```python
plt.scatter(x, siny)
plt.scatter(x, cosy)
plt.show()
```
